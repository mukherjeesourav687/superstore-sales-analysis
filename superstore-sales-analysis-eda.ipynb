# Superstore Sales Data Analysis

## Objective
The goal of this analysis is to explore the Superstore dataset and understand sales patterns across different product categories.

## Tools Used
- Python
- Pandas
- Matplotlib

## Key Analysis
- Examined the structure of the dataset using df.info()
- Performed exploratory data analysis using df.describe()
- Calculated total sales by product category
- Visualized category-wise sales using a bar chart

## Insights
- Technology category generates the highest sales.
- Office Supplies contributes a moderate share of revenue.
- Furniture has comparatively lower sales.

This analysis provides a quick overview of category performance in the Superstore dataset.

```python
import pandas as pd
import os

for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        path = os.path.join(dirname, filename)

df = pd.read_csv(path)

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sales_by_category = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)

sales_by_category
```

```python
import matplotlib.pyplot as plt

sales_by_category.plot(kind='bar')

plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.show()
```

Insights

Technology category generates the highest sales in the Superstore dataset.
Office Supplies contributes a moderate share of revenue.
Furniture generates comparatively lower sales.

This analysis helps understand category-wise sales performance.

```python
sales_by_region = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)

sales_by_region
```

```python
sales_by_region.plot(kind="bar")
plt.title("Sales by Region")
plt.show()
```
